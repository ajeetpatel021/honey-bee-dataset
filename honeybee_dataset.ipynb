```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('honeyproduction.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
prod_per_year = df.groupby('year')
```

```python
prod_per_year.head()
```

```python
prod_per_year = df.groupby('year').totalprod.mean().reset_index()
```

```python
prod_per_year.head()
```

```python
X = prod_per_year['year']
print(X)
```

```python
X.shape
```

```python
X = X.values.reshape(-1,1)
```

```python
print(X.shape)
print(X)
```

```python
X.shape
```

```python
Y = prod_per_year['totalprod']
```

```python
Y.shape
```

```python
Y = Y.values.reshape(-1,1)
```

```python
Y.shape
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
```

```python
regr.fit(X,Y)
```

```python
print("slope = {}, intercept = {}".format(regr.coef_,  regr.intercept_))
```

```python
y_predictions = regr.predict(X)
```

```python
plt.scatter(X,Y)
plt.plot(X,y_predictions)
plt.show()
```

```python
X_future = np.array(range(2013,2031))
```

```python
X_future = X_future.reshape(-1,1)
```
